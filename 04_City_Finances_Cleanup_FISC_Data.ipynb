```python
#GeoPy times out on all records in the fisc_clean_df, so creating this dataframe of each city, state
city_state_df = fisc_clean_df[['city_name','state']].drop_duplicates().reset_index()

city_state_df = city_state_df.drop('index', axis = 'columns')
```

```python
#Get the latitude and longitude of each city, state
geolocator = Nominatim()

def lat_long(row):
    str_address = row.city_name + ', ' + row.state
    location = geolocator.geocode(str_address, country_codes = 'US', timeout = 100)

    if location != None:
        row['latitude'] = location.latitude
        row['longitude'] = location.longitude
    
    return row
    
city_state_df = city_state_df.apply(lat_long, axis = 'columns')
```

```python
fisc_clean_df = pd.merge(fisc_clean_df, city_state_df, on =['city_name','state'])
```

```python
out_csv = 'data/02_fisc_cleaned_data.csv'

fisc_clean_df.to_csv(out_csv, index=None)
```
