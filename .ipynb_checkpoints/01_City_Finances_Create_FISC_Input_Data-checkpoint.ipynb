```python
#Imports
import pandas as pd
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
#Set options
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
#Open spreadsheet
all_data_df = pd.read_excel('Data/fisc_full_dataset_2016_update.xlsx', sheetname='Dataset')

all_data_df.sample()
```

```python
#Create dataframe of core and FISC columns
core_df = all_data_df.iloc[:, 0:5]
```

```python
rev_and_expense_df = all_data_df.iloc[:,525:]
```

```python
fisc_df = core_df.join(rev_and_expense_df)
```

```python
#Include the last 15 years of data in this project
fisc_df = fisc_df[fisc_df.year > 2001]
```

```python
fisc_df.head(2)
```

```python
def split_state_city(row):
    state_city = str(row.city_name)
    
    row['state'] = state_city[0:2]
    
    row['city_name'] = state_city[4:]
    
    return row

fisc_df = fisc_df.apply(split_state_city, axis='columns')

fisc_df.head(2)
```

```python
#Write out the fisc dataframe to a csv file
out_csv = 'Data/01_fisc_input_data.csv'

fisc_df.to_csv(out_csv, index = False)
```
